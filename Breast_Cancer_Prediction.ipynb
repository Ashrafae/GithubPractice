## Dataset Information/Citation

The dataset provides valuable insights into breast cancer diagnosis, as it contains features computed from digitized images of fine needle aspirates (FNA) of breast masses. These features describe characteristics of cell nuclei present in the images.

The separation between malignant and benign tumors was determined using the Multisurface Method-Tree (MSM-T), a classification technique that leverages linear programming to construct a decision tree. Relevant features were selected through an exhaustive search, and the resulting linear program was employed to obtain the separating plane in a 3-dimensional space.

This dataset is accessible via the UCI Machine Learning Repository and has been widely used for research and analysis:

- Wolberg, William; Mangasarian, Olvi; Street, Nick; and Street, W. (1995). Breast Cancer Wisconsin (Diagnostic). UCI Machine Learning Repository. [DOI](https://doi.org/10.24432/C5DW2B).

# Taking a closer look at the feature 'texture error'. Should the feature be dropped or not?

```python
sns.violinplot(x='outcome', y='texture error', data=df)
plt.title('Texture Error vs Outcome')
plt.show()
```

The visualization of a violin plot and the results of a T-Test conducted in the preceding analysis suggest the possibility of dropping the 'texture error' feature. However, it's important to consider the context in which this decision is made. In this particular scenario, I opt to retain the 'texture error' feature despite these indications for several reasons. Firstly, the dataset is relatively small, and losing data can significantly impact the model's ability to learn patterns effectively. Additionally, the 'texture error' feature may contain valuable information, even if it appears less influential in isolation. Hence, for the sake of preserving valuable information and ensuring the robustness of othe model, I choose to retain 'texture error' as a feature in our dataset.

```python
from scipy import stats

outcome_column = 'outcome'
feature_column = 'texture error'

# Calculate the mean texture error for each outcome category
mean_texture_error_by_outcome = df.groupby(outcome_column)[feature_column].mean()

# Display the means for each outcome
print("Mean Texture Error by Outcome:")
print(mean_texture_error_by_outcome)

# Perform a t-test to assess if the means are significantly different
outcome_0 = df[df[outcome_column] == 0][feature_column]
outcome_1 = df[df[outcome_column] == 1][feature_column]

t_statistic, p_value = stats.ttest_ind(outcome_0, outcome_1)

print("\nT-Test Results:")
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

# Interpret the p-value
alpha = 0.05  # significance level
if p_value < alpha:
    print("\nThe means are significantly different.")
else:
    print("\nThe means are not significantly different.")
```

Logistic regression is a statistical model that in its basic form uses a logistic function to model a binary dependent variable, although many more complex extensions exist.
